```python
# I'm just going to manually type in the data...


import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import json
import seaborn as sns
sns.set()

"""ORDER of measurements IS 200, 100, 50, 20, 5, 3, 1"""

results_trained_on_200 = [0.47066, 0.43413, 0.39241, 0.34533, 0.33456, 0.34988, 0.392]
results_trained_on_100 = [0.47041, 0.43395, 0.39214, 0.34517, 0.33402, 0.34985, 0.3921]
results_trained_on_50 = [0.47064, 0.43423, 0.39241, 0.34548, 0.3349, 0.35025, 0.393]
results_trained_on_20 = [0.47037, 0.43389, 0.39216, 0.34522, 0.33411, 0.34988, 0.3926]
results_trained_on_5 = [0.47051, 0.4341, 0.39223, 0.34556, 0.33441, 0.35002, 0.3937]
results_trained_on_3 = [0.47, 0.43344, 0.39185, 0.34444, 0.33298, 0.34947, 0.3928]
results_trained_on_1 = [0.47055, 0.43412, 0.39222, 0.3454, 0.33447, 0.35009, 0.3946]
```

```python
plt.clf()

plt.plot(range(7), results_trained_on_200, color='red')
plt.plot(range(7), results_trained_on_100, color='orange')
plt.plot(range(7), results_trained_on_50, color='yellow')
plt.plot(range(7), results_trained_on_20, color='green')
plt.plot(range(7), results_trained_on_5, color='blue')
plt.plot(range(7), results_trained_on_3, color='black')
plt.plot(range(7), results_trained_on_1, color='purple')
plt.xticks(range(7), ['200', '100', '50', '20', '5', '3', '1'])
plt.show()
```

```python
# They're all like exactly the same... so let's look at deviation from mean...

results_trained_on_200 = np.asarray(results_trained_on_200)
results_trained_on_100 = np.asarray(results_trained_on_100)
results_trained_on_50 = np.asarray(results_trained_on_50)
results_trained_on_20 = np.asarray(results_trained_on_20)
results_trained_on_5 = np.asarray(results_trained_on_5)
results_trained_on_3 = np.asarray(results_trained_on_3)
results_trained_on_1 = np.asarray(results_trained_on_1)

mean_results = (results_trained_on_200 +
                results_trained_on_100 +
                results_trained_on_50 +
                results_trained_on_20 +
                results_trained_on_5 +
                results_trained_on_3 +
                results_trained_on_1) / 7


```

```python
plt.plot(range(7), results_trained_on_200 - mean_results, color='red', label="trained on 200")
plt.plot(range(7), results_trained_on_100 - mean_results, color='orange', label="trained on 100")
plt.plot(range(7), results_trained_on_50 - mean_results, color='yellow', label="trained on 50")
plt.plot(range(7), results_trained_on_20 - mean_results, color='green', label="trained on 20")
plt.plot(range(7), results_trained_on_5 - mean_results, color='blue', label="trained on 5")
plt.plot(range(7), results_trained_on_3 - mean_results, color='black', label="trained on 3")
plt.plot(range(7), results_trained_on_1 - mean_results, color='purple', label="trained on 1")

leg = plt.legend(fontsize=10, shadow=True, loc=(0.05, 0.6))

plt.xticks(range(7), ['200', '100', '50', '20', '5', '3', '1'])
plt.show()
```
